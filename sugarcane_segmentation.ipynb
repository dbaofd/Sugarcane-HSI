```python
"""
This script is used for sugarcane semantic label generation. For a given subdataset, 
All the images inside will have their corresponding semantic label image generated. 

Parameters need change:
1) Second block, change the path and name of the model.
2) SOURCE_PATH
3) SAVING_PATH
"""

import numpy as np
from spectral import *
from PIL import Image
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
from sklearn import metrics
import pickle
import os
import math
from skimage import morphology
import skimage
HEIGHT=1088
WIDTH=2048
BAND=161
```

```python
# Load from file
os.chdir("C:\imecdata")
with open("svm_model_improved_5.pkl", 'rb') as file:
    model = pickle.load(file)
```

```python
dir_list = []
file_list = []
count = 0
total_number_of_images=0
# Load the calibrated hsi dataset
SOURCE_PATH='E:\Sugarcane_Dataset_Normalized\sugarcane_1_Sep_2021_Norm'# subdataset path.
SAVING_PATH='E:\Sugarcane_Dataset_Label\Sep1'# path to save the labels.
for (root, dirs, files) in os.walk(SOURCE_PATH, topdown=True):
    if len(dirs) != 0:
        dir_list = dirs
    if count > 0:
        li = []
        for i in files:
            if i.endswith('.hdr'):
                li.append(i)
                total_number_of_images+=1
        file_list.append(li)
    count += 1

count = 0
#li=["Q205", "Q208", "CP29-116","Nco310","Q44","Q68","Q78","Q82","Q124","Q171"]
# Target folders
li=["CP29-116","Nco310","Q68", "Q82","Q44","Q124","Q171","Q205", "Q208","Q78"]
for i in range(len(dir_list)):#len(dir_list)
#     if dir_list[i] in li:
    for j in range(len(file_list[i])):#len(file_list[i])
        count+=1
        print("*********************************{}/{}**************************************".format(count, total_number_of_images))
        os.chdir(SOURCE_PATH)
        img_path = dir_list[i] + "/" + file_list[i][j]

        print(img_path)
        img=envi.open(img_path)[:,:,:]
        img=np.reshape(img, (-1, BAND)) # Reshape the image array so it can be fed into svm model.
        predicted_label=model.predict(img)
        predicted_label=np.reshape(predicted_label,(HEIGHT,WIDTH))
        new_label=np.zeros((HEIGHT, WIDTH, 3)) # Create label matrix. 
        for k in range(HEIGHT):
            for t in range(WIDTH):
                if predicted_label[k][t]==1:
                    new_label[k][t][0]=128
        rescale_im=(255.0/new_label.max()*(new_label-new_label.min())).astype(np.uint8)
        im2=Image.fromarray(rescale_im)
        os.chdir(SAVING_PATH)
        save_path = dir_list[i] + "/" + file_list[i][j].split(".")[0]+".png"
        #save_path = file_list[i][j].split(".")[0]+".png"
        print(save_path)
        im2.save(save_path)
```
